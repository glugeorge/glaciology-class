```python
# time domain
dt = 10                       # time step, units [years]
T = 3000                          # total length of simulation, units [years]
t = np.linspace(0,T,round(T/dt))  # the time grid, units [years]
Lt = t.size                       # record the length of the time grid for use later
```

```python
# space domain
dx = 200                          # grid spacing, units [m]
X = 40000                         # domain length, units [m]
x = np.linspace(0,X,round(X/dx))  # spatial grid, units [m]
Lx = x.size                       # record the length of the spatial grid for use later
```

```python
n = 3.0                              # The flow law exponent
seconds_per_year = 365*24*60*60      # approximate number of seconds in year
A = 24*10**(-25) * seconds_per_year  # The flow law parameter, units [Pa a] (original value from Cuffey and Paterson 24e-25, units Pa s]
rho = 917                            # ice density,  units [kg/m^3]
g = 10                               # acceleration due to gravity, units [m/s^2]
e = 0.0001                           # used to prevent ice thickness from reaching zero
```

Define equations to solve:


```python
def to_solve(var):
    h = var[0:Lx]
    h_old = h
    q = var[Lx:]
    q[-1] = -(2*A)/(n+2) * (rho*g)**n * h_old[0]**(n+2) * ((h_old[-1] - h_old[-2])/dx)**n - q[-1] # use one sided difference instead of centered
    q[0] = q[0] 
    q[1:-1] = -(2*A)/(n+2) * (rho*g)**n * h_old[1:-1]**(n+2) * ((h_old[2:]-h_old[0:-2])/(2*dx))**n - q[1:-1]
    
    h[0] = h_old[0] - h[0] + dt*accum(x[0]) # use one sided difference instead of centered
    h[-1] = h_old[-1] - h[-1] + dt*(accum(x[-1]) - (q[-1]-q[-2])/dx) # use one sided difference instead of centered
    h[1:-1] = h_old[1:-1] - h[1:-1] + dt*(accum(x[1:-1]) - (q[2:]-q[0:-2])/dx)
    return np.concatenate((h,q))
    
def accum(x):
    return 1e-4*(X/3 -x)
    
```

```python
result_h = np.zeros((Lt,Lx))
result_q = np.zeros((Lt,Lx))
init_h = np.ones(x.shape)*e
init_q = np.zeros(x.shape)
init_q[-1] = 0
var = [init_h, init_q]
for i in range(len(t)):
    if i%100 == 0:  
        print(f't={t[i]}')
    res = sp.optimize.fsolve(to_solve,var)
    result_h[i] = res[0:Lx]
    result_h[i][result_h[i]<e] = e
    result_q[i] = res[Lx:]
    var = res
```
