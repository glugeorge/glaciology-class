# Practical 1: Data analysis and solving equations in python

```python
# Load relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

## How many data points are in `SIR_density.csv`?

```python
data = pd.read_csv('Practical 1/SIR_density.csv')
```

```python
depth = data['depth [m]']
density = data['density [kg/m^3]']
```

```python
print(f"There are {len(data)} data points")
```

## How deep is the ice core?

```python
print(f"The max depth of the core is {max(depth)} m")
```

## What is the density at the surface? What is the density at 30 m?

```python
density_0_30 = np.interp([0,30],depth,density)
```

```python
print(f"The density at the surface is {density_0_30[0]} kg/m^3 and {density_0_30[1]} kg/m^3 at 30 m")

```

## Plot density against depth.

```python
plt.plot(depth,density,'.')
plt.xlabel('depth [m]')
plt.ylabel('density [kg/m^3]')
```

## What is the average density in the top 50 m and in the bottom 200m? 

```python
density_50 = density[[i for i in range(len(density)) if depth[i]<50]]
density_200 = density[[i for i in range(len(density)) if depth[i]>200]]
```

```python
print(f"The average density in the top 50 m is {np.mean(density_50)} kg/m^3, and in the bottom 200 m it is {np.mean(density_200)} kg/m^3.")
```

## Compute and plot pressure as a function of depth P(z).

```python
P = np.zeros(depth.shape)
for i in range(1,len(depth)):
    if i == 0:
        P[i] = 0
    else:
        dz = depth[i]-depth[i-1]
        P[i] = P[i-1]+dz*density[i]*9.81
```

```python
plt.plot(depth,P)
plt.xlabel("depth [m]")
plt.ylabel("pressure [Pa]")
```

## Indicate on the plot the approximate pressure a human hand is capable of producing.
Assuming 50 to 100 kPa 

```python
plt.plot(depth,P)
plt.fill_between(np.arange(0,40),50*1000*np.ones(40),100*1000*np.ones(40), alpha=0.5)
plt.text(40,50*1000,'range of human hand pressures')
plt.xlabel("depth [m]")
plt.ylabel("pressure [Pa]")
```

## When computing P in ice sheet models they usually use 𝑃 = 𝜌𝑔𝑧 and assume a constant density 𝜌 = 918 kg/m3. How inaccurate is this approximation?

```python
P_const = 918*9.81*depth
```

```python
plt.plot(depth,P,label='Pressures with variable density')
plt.plot(depth,P_const,label='Pressures with constant density')
plt.xlabel("depth [m]")
plt.ylabel("pressure [Pa]")
plt.legend()
```

It is inaccurate at the surface, but then the slope becomes the same at greater depths

## How would your answer change if I told you the data do not reach the ice base and the ice is actually 660m thick in this location?

I would use the answer with the density data, but then use the constant data to calculate the pressures at greater depths. I still however wont make the assumption for the entire profile because the offset caused by variable densities at the surface propogates to all the values

```python
depths_660 = np.append(depth.to_numpy(),np.arange(300,661))
densities_660 = np.append(density.to_numpy(),918*np.ones(np.arange(300,661).shape))
P_660 = np.zeros(depths_660.shape)

for i in range(1,len(depths_660)):
    if i == 0:
        P_660[i] = 0
    else:
        dz = depths_660[i]-depths_660[i-1]
        P_660[i] = P_660[i-1]+dz*densities_660[i]*9.81

P_const_660 = 918*9.81*depths_660
plt.plot(depths_660,P_660,label='Pressures with variable density')
plt.plot(depths_660,P_const_660,label='Pressures with constant density')
plt.xlabel("depth [m]")
plt.ylabel("pressure [Pa]")
plt.legend()
```
